# Data Visualization Assignment

## Introducción

En este trabajo se presentan diferentes tipos de visualizaciones de datos utilizando las librerías Matplotlib y Seaborn en Python.

Cada gráfico fue creado para representar distintos tipos de información y cada visualización incluye personalización de colores, estilos y etiquetas.

```python
# =========================
# IMPORTACIONES
# =========================

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Crear carpeta para guardar imágenes
os.makedirs("imgs", exist_ok=True)

plt.style.use('seaborn-v0_8')

# Semilla aleatoria para tener siempre los mismos gráficos.
np.random.seed(42)
```

# 1. Gráfico de Línea

Este gráfico muestra cómo cambia la temperatura promedio a lo largo de los meses del año.

Elegí un gráfico de línea porque es una de las mejores formas de representar tendencias y cambios en datos temporales.

```python
# Datos
meses = np.arange(1, 13)

temperaturas = [22, 24, 27, 30, 32, 35, 37, 36, 33, 29, 25, 23]

plt.figure(figsize=(8,5))

# Grafico
plt.plot(
    meses,
    temperaturas,
    color='blue',
    marker='o',
    linestyle='--',
    linewidth=2,
    label='Temperatura'
)

# Título y etiquetas
plt.title('Temperatura Promedio Mensual')
plt.xlabel('Mes')
plt.ylabel('Temperatura (°C)')

plt.legend()

# Guarda la imagen
plt.savefig('imgs/grafico_linea.png', dpi=300, bbox_inches='tight')

# Mostrar gráfico
plt.show()
```

# 2. Gráfico de Barras

Este gráfico compara las notas obtenidas en distintas materias.

El gráfico de barras es útil para comparar categorías diferentes y observar rápidamente cuáles tienen valores más altos o más bajos.

```python
# Datos
materias = ['Matemática', 'Física', 'Química', 'Biología', 'Historia']

notas = [88, 92, 79, 85, 90]

plt.figure(figsize=(8,5))

# Crea gráfico de barras
plt.bar(
    materias,
    notas,
    color=['red', 'blue', 'green', 'orange', 'purple']
)

# Título y etiquetas
plt.title('Notas por Materia')
plt.xlabel('Materias')
plt.ylabel('Nota (%)')

# Guarda la imagen
plt.savefig('imgs/grafico_barras.png', dpi=300, bbox_inches='tight')

# Mostrar gráfico
plt.show()
```

# 3. Histograma

Este histograma representa la distribución de puntajes de un grupo de estudiantes.

Elegí esta visualización porque permite observar cómo se distribuyen los datos y detectar concentraciones o variaciones en los valores.

```python
# Datos
puntajes = np.random.normal(75, 10, 1000)

plt.figure(figsize=(8,5))

# Crear histograma
plt.hist(
    puntajes,
    bins=20,
    color='skyblue',
    edgecolor='black'
)

# Título y etiquetas
plt.title('Distribución de Puntajes')
plt.xlabel('Puntaje')
plt.ylabel('Frecuencia')

# Guarda la imagen
plt.savefig('imgs/histograma.png', dpi=300, bbox_inches='tight')

# Mostrar gráfico
plt.show()
```

# 4. Gráfico de Dispersión

Este gráfico muestra la relación entre las horas de estudio y los resultados de un examen.

El gráfico de dispersión es apropiado para visualizar relaciones o correlaciones entre dos variables numéricas.

```python
# Datos
horas_estudio = np.random.randint(1, 10, 50)

resultados = horas_estudio * 8 + np.random.normal(0, 5, 50)

plt.figure(figsize=(8,5))

# Crea el  gráfico de dispersion
plt.scatter(
    horas_estudio,
    resultados,
    color='darkgreen',
    marker='o',
    label='Estudiantes'
)

# Titulo y etiquetas
plt.title('Horas de Estudio vs Resultado')
plt.xlabel('Horas de Estudio')
plt.ylabel('Resultado del Examen')

plt.legend()

# Guarda la imagen
plt.savefig('imgs/grafico_dispersion.png', dpi=300, bbox_inches='tight')

# Mostrar gráfico
plt.show()
```

# 5. Heatmap

Este mapa de calor compara distintas laptops según la calidad de pantalla, teclado y rendimiento.

Elegí esta visualización porque permite comparar múltiples categorías rápidamente utilizando colores para representar los puntajes.

```python
# Datos de laptops
datos = np.array([
    [9, 8, 10],   # Laptop 1
    [7, 9, 8],    # Laptop 2
    [8, 7, 9]     # Laptop 3
])

plt.figure(figsize=(8,6))

# Crea Heatmap
sns.heatmap(
    datos,
    annot=True,
    cmap='coolwarm',
    linewidths=0.5
)

# Etiquetas
plt.xticks(
    [0.5,1.5,2.5],
    ['Pantalla', 'Teclado', 'Rendimiento']
)

plt.yticks(
    [0.5,1.5,2.5],
    ['Laptop A', 'Laptop B', 'Laptop C'],
    rotation=0
)

# Titulo
plt.title('Comparación de Laptops')

# Guarda la imagen
plt.savefig('imgs/heatmap.png', dpi=300, bbox_inches='tight')

# Muestra el grafico
plt.show()
```
